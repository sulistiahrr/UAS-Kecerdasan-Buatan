```python
import matplotlib.pyplot as plt
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
kualitas_makanan = ctrl.Antecedent(np.arange(0, 11, 1), 'kualitas makanan')
kecepatan_pelayanan = ctrl.Antecedent(np.arange(0, 11, 1), 'kecepatan pelayanan')
suasana_restoran = ctrl.Antecedent(np.arange(0, 11, 1), 'suasana restoran')
```

```python
kualitas_makanan['Tidak Lezat'] = fuzz.trapmf(kualitas_makanan.universe, [0, 0, 1, 5])
kualitas_makanan['Lezat'] = fuzz.trapmf(kualitas_makanan.universe, [5, 9, 10,10])
```

```python
kecepatan_pelayanan['Jelek'] = fuzz.trimf(kecepatan_pelayanan.universe, [0, 0, 4])
kecepatan_pelayanan['Sedang'] = fuzz.trimf(kecepatan_pelayanan.universe, [1, 5, 9])
kecepatan_pelayanan['Bagus'] = fuzz.trimf(kecepatan_pelayanan.universe, [6, 10, 10])
```

```python
suasana_restoran['Santai'] = fuzz.trimf(suasana_restoran.universe, [0, 5, 10])
suasana_restoran['Klasik'] = fuzz.trimf(suasana_restoran.universe, [10, 15, 20])
suasana_restoran['Pedesaan'] = fuzz.trimf(suasana_restoran.universe, [20, 25, 30])
```

```python
kualitas_makanan.view()
kecepatan_pelayanan.view()
suasana_restoran.view()
```
